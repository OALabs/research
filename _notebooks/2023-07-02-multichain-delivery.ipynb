## Overview
We are going to analyze a multi-stage delivery chain that ends up delivering [AsyncRAT](https://www.unpac.me/results/843ac4a3-f058-4a1a-9eda-ead3f0c2d741#/). This delivery chain has multiple stages that are responsible for preparing the target host for the delivery of the malware such that the host will likely not detect the final payload.
